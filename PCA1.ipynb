Q1. What are Eigenvalues and Eigenvectors? How are they related to the Eigen-Decomposition approach?

Eigenvalues and eigenvectors are concepts from linear algebra that are associated with square matrices. Let's break down these terms:

Eigenvalues (λ): These are scalar values that represent how a linear transformation, represented by a matrix, scales a vector. In simpler terms, an eigenvalue tells you how much a matrix scales or stretches a vector. If A is a square matrix and v is an eigenvector of A, then the product Av is a scaled version of v, and the scaling factor is the eigenvalue λ.

Mathematically, for a matrix A and an eigenvector v:

A * v = λ * v

Eigenvectors (v): Eigenvectors are non-zero vectors that only get scaled by a matrix; their direction doesn't change. In the equation mentioned above, v is the eigenvector corresponding to the eigenvalue λ.
